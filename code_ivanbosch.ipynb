```python
maxvalues=[]

years=np.array((n00,n01,n02,n03,n04,n05,n06,n07,n08,n09,n10,n11,n12,n13,n14,n15,n16,n17,n18,n19,n20,n21)) #numpyarray of nparrays

for year in years:
    H,X1 = np.histogram( year, bins = 100000000, normed = False ) #this amount of bins doesn't has mathematical sense but provides us more accuracy for the max value
    maxvalue=np.amax(H)

    ind=np.where(H==maxvalue)
    ind=np.array(ind)
    ind=ind.tolist()

    cont=0
    suma=0.
    for i in ind:
        for j in i:
            cont=cont+1
            suma=suma+X1[j] #if there's two bins with the same count of the hist we compute the mean as maxvalue
    popmax=suma/cont
    maxvalues.append(popmax)

maxvalues=np.array(maxvalues)
xyears=np.arange(2000,2022,1)

plt.plot(xyears,maxvalues,'o-',linewidth=0.5,c='red', label='Max. of pop.')

modelmaxs=np.polyfit(xyears,maxvalues, 1)

f1=modelmaxs[0]*xyears+modelmaxs[1] #lineal regression

plt.grid()
plt.ylabel('Pop. with max. num. of municipies')
plt.xlabel('YEAR')
plt.title('Maximum of the distribution')

plt.plot(xyears,f1,c='black',label='Lineal regression')

plt.legend(loc='upper center')

plt.savefig('xm.png')            
```

```python
alphas=[]

for year in years:
    H,X1 = np.histogram( year, bins = 85, normed = True ) #more bins means more fluctuations in the histogram, for the regression is better a smoother behaviour
    
    dx=X1[1]-X1[0]
    xdeve=X1[1:]-dx/2.
    
    logH=np.log(H)
    logX1=np.log(xdeve)


    model = np.polyfit(logX1,logH, 1)

    alphas.append(model[0]) #slope of the linear regression
      
alphas=np.array(alphas)
alphas=-(alphas[:]+1.)  #alpha of the Pareto distr. expression
alphaslin=np.delete(alphas,3)
xyearslin=np.delete(xyears,3)

modelalphas=np.polyfit(xyearslin,alphaslin, 1)

falpha=modelalphas[0]*xyearslin+modelalphas[1]

plt.rcParams['text.usetex'] = True
plt.grid()
plt.ylabel(r'$\alpha$')
plt.xlabel('YEAR')
plt.title('Alpha coefficient for each year')


plt.plot(xyears,alphas,'o-',linewidth=0.5,c='red',label=r'$\alpha$')
plt.plot(xyearslin,falpha,c='black',label='Linear regression')

plt.legend(loc='upper right')

plt.savefig('alpha.png')

    
```

```python
x21=np.arange(maxvalues[21],2000,1)
pareto21=(alphas[21]*maxvalues[21]**alphas[21])/(x21**(alphas[21]+1))
x00=np.arange(maxvalues[0],2000,1)
pareto00=(alphas[0]*maxvalues[0]**alphas[0])/(x00**(alphas[0]+1))

plt.hist(n21, bins=300, density=True,color='blue',linewidth=2,alpha=0.4,lw=2,ls='dashed',label='Pop. histogram y=2021')
plt.hist(n00, bins=300, density=True,color='r',linewidth=2,alpha=0.3,lw=2,ls='dashed',label='Pop. histogram y=2000')
plt.plot(x21,pareto21,c='blue',lw=3,label='Pareto fit y=2021')
plt.plot(x00,pareto00,c='red',lw=3, label='Pareto fit y=2000')

plt.xlabel('Population')
plt.ylabel('f(p)')
plt.xlim([maxvalues[21],2000])
plt.ylim([0,0.0025])
plt.grid()
plt.legend()
plt.savefig('2100.png')
```
